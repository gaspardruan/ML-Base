```python
import pandas as pd
import numpy as np

data = pd.read_csv('iris_data.csv')
data.head()
```

```python
x = data.drop(['target', 'label'], axis=1)
y = data.loc[:, 'label']
```

```python
from sklearn import tree

dt = tree.DecisionTreeClassifier(criterion='entropy', min_samples_leaf=5)
dt.fit(x, y)
```

```python
from sklearn.metrics import accuracy_score
y_predict = dt.predict(x)
accuracy = accuracy_score(y, y_predict)
print(accuracy)
```

```python
from matplotlib import pyplot as plt

fig = plt.figure(figsize=(10, 10))
tree.plot_tree(dt, filled=True, feature_names=x.columns, class_names=['setosa', 'versicolor', 'virginica'])
```

```python
dt2 = tree.DecisionTreeClassifier(criterion='entropy', min_samples_leaf=1)
dt2.fit(x, y)
```

```python
fig2 = plt.figure(figsize=(10, 10))
tree.plot_tree(dt2, filled=True, feature_names=x.columns, class_names=['setosa', 'versicolor', 'virginica'])
```
